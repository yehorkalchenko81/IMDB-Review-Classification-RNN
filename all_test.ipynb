```python
import matplotlib.pyplot as plt
from keras.datasets import imdb
from keras.utils import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense, Bidirectional
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from keras import optimizers
```

```python
max_features = 10000
maxlen = 500
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
```

```python
x_train = pad_sequences(x_train, maxlen=maxlen)
x_test = pad_sequences(x_test, maxlen=maxlen)
```

# SimpleRNN, LSTM, Bidirectional

```python
for model_type in [SimpleRNN, LSTM, Bidirectional]:

    model = Sequential()

    model.add(Embedding(max_features, 32))

    if model_type == Bidirectional:
        model.add(model_type(LSTM(32)))
    else:
        model.add(model_type(32))

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=optimizers.RMSprop(0.001), loss='binary_crossentropy', metrics=['acc'])

    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.2, min_lr=0.0001)

    print(f'Model_Type: {model_type}')

    history = model.fit(x_train, y_train, epochs=15, batch_size=64, callbacks=[early_stop, reduce_lr], validation_data=(x_test, y_test))

    plt.figure(figsize=(13, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.title('Loss')

    plt.subplot(1, 2, 2)
    plt.plot([round(100*e, 2) for e in history.history['acc']], label='train')
    plt.plot([round (100*e, 2) for e in history.history['val_acc']], label='test')
    plt.title('Accuracy')

    plt.ylim(0, 100)
    plt.legend()
    plt.tight_layout()
    plt.show()

    predictions = model.predict(x_test)

    binary_predictions = [1 if pred > 0.5 else 0 for pred in predictions]

    for i in range(10):
        print(f"Прогнозоване: {binary_predictions[i]}, Фактичне: {y_test[i]}")

    test_loss, test_acc = model.evaluate(x_test, y_test)
    print('Test Loss: {}'.format(test_loss))
    print('Test Accuracy: {}'.format(test_acc))

    print(200*'_')
    
```

# Deep RNN

```python
model = Sequential()

model.add(Embedding(max_features, 32))

model.add(Bidirectional(LSTM(32, return_sequences=True)))
model.add(Bidirectional(LSTM(32)))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer=optimizers.RMSprop(0.001), loss='binary_crossentropy', metrics=['acc'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.2, min_lr=0.0001)

history = model.fit(x_train, y_train, epochs=15, batch_size=64, callbacks=[early_stop, reduce_lr], validation_data=(x_test, y_test))
```

```python
plt.figure(figsize=(13, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.title('Loss')

plt.subplot(1, 2, 2)
plt.plot([round(100*e, 2) for e in history.history['acc']], label='train')
plt.plot([round (100*e, 2) for e in history.history['val_acc']], label='test')
plt.title('Accuracy')

plt.ylim(0, 100)
plt.legend()
plt.tight_layout()
plt.show()

predictions = model.predict(x_test)

binary_predictions = [1 if pred > 0.5 else 0 for pred in predictions]

for i in range(10):
    print(f"Прогнозоване: {binary_predictions[i]}, Фактичне: {y_test[i]}")


test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test Loss: {}'.format(test_loss))
print('Test Accuracy: {}'.format(test_acc))
```
